```python
import pandas as pd
import numpy as np
```

```python
n_users = 500
```

```python
np.random.seed(42)

user_id = np.arange(1, n_users + 1)

login_count_30d = np.random.poisson(lam=10, size=n_users)

session_duration_avg = np.random.normal(loc=10, scale=5, size=n_users)
session_duration_avg = np.clip(session_duration_avg, 1, 60)

features_used = np.random.randint(1, 10, size=n_users)

days_since_last_login = np.random.randint(0, 30, size=n_users)
```

```python
churn_probability = (
    0.4 * (days_since_last_login / 30) +
    0.3 * (1 - login_count_30d / max(login_count_30d)) +
    0.3 * (1 - features_used / max(features_used))
)

churned = np.random.binomial(1, churn_probability)
```

```python
df = pd.DataFrame({
    "user_id": user_id,
    "login_count_30d": login_count_30d,
    "session_duration_avg": session_duration_avg,
    "features_used": features_used,
    "days_since_last_login": days_since_last_login,
    "churned": churned
})

df.head()
```

```python
df.to_csv("user_activity.csv", index=False)
```
